### Droping irrelevant data from the dataset

```python
df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],inplace=True,axis=1)
```

```python
df.head()
```

```python
df['Date']=pd.to_datetime(df['Date'],format='%Y-%m-%d')
```

```python
df.head()
```

```python
## Active Cases
```

```python
df['Active_Cases']=df['Confirmed']-(df['Cured']+df['Deaths'])
```

```python
df.tail()
```

```python
statewise = pd.pivot_table(df,values=['Confirmed','Deaths','Cured'],index='State/UnionTerritory',aggfunc=max)
```

```python
statewise['Recovery_Rate']=statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise['Moratality_Rate']=statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise=statewise.sort_values(by='Confirmed',ascending=False)
```

```python
statewise.style.background_gradient(cmap='cubehelix')
```

```python
# Top 10 Active Cases States
```

```python
ActiveCases10=df.groupby(by='State/UnionTerritory').max()[['Active_Cases','Date']].sort_values(by=['Active_Cases'],ascending=False).reset_index()
```

```python
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 States with most active cases in INDIA")
ax=sns.barplot(data=ActiveCases10.iloc[:10],y='Active_Cases',x='State/UnionTerritory',linewidth=2,edgecolor='red')
```

```python
# Top States with Highest Deaths
```

```python
top_10_deaths=df.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths'],ascending=False).reset_index()
```

```python
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 States with Highest Deaths in INDIA")
ax=sns.barplot(data=top_10_deaths.iloc[:12],y='Deaths',x='State/UnionTerritory',linewidth=2,edgecolor='red')
```
